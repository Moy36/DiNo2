```python
def intv_EF02_L3(intv_col, attri_list, name_list):
    count = len(attri_list)
    
    for column in name_list:
        interview_L123[column] = pd.Series([np.nan] * len(interview_L123), dtype='object')
    
    for i in range(count):
        for index, value in intv_col.items():
            if isinstance(value, str) and pd.notna(value):
                items = value.split(',')
                
                for item in items:
                    item = item.strip()
                    
                    if item in attri_list[i]:
                        current_value = interview_L123.loc[index, name_list[i]]
                        
                        if pd.isna(current_value):
                            interview_L123.loc[index, name_list[i]] = item
                        else:
                            interview_L123.loc[index, name_list[i]] = f"{current_value}, {item}"
    
    return interview_L123
```
