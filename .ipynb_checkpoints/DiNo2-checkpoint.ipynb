Calculates the total score of level1 attributes, and the probability of each level1 question occurring (Here, we separate the multiple questions within each block's level1. Even though we have set a rule to randomly select three questions, it is still necessary to calculate the probabilities because the random selection could result in uneven coverage of questions, potentially skewing the overall results.) Finally, computes the normalized weights for each level1 attributess based on the total score and the probability.

```python
Whole_L1.to_csv('Whole_L1.csv')
total_score = Whole_L1[level1_new].sum()
print(total_score)
answer_count = Whole_L1[level1_new].notnull().sum()
total_count = answer_count.sum()
L1_probability = round(answer_count/total_count, 3)
print(L1_probability)
Weight_L1 = round((total_score/L1_probability)/sum(total_score/L1_probability),3)

L1_weight = dict(Weight_L1)
print(L1_weight)
print(sum(L1_weight.values()))
```

Defines a function 'count_key' to count the occurrences of each element in the specified column (col) of each block.

```python
def count_key(block, col):
    count_dict = {}
    for value in block[col]:
        if pd.notna(value):
            items = value.split(',')
         
            for item in items:
                item = item.strip()
                if item in count_dict:
                    count_dict[item] += 1
                else:
                    count_dict[item] = 1
    return count_dict
```

Defines a function 'sub_weight' to calculate the weights of each element that affect level1 attributes in each block.

```python
def sub_weight(dic):
    weight_dic = {}
    total= sum(dic.values())
    for key in dic:
        weight_dic[key]=round(dic[key]/total,3)
    return weight_dic
    
```

Defines a function that preprocess quantitative survey questions.

```python
def prefunc(data,dic):
    data.rename(columns=dic,inplace=True)
    for col in data.columns:
        data[col]=pd.to_numeric(data[col],errors='coerce').astype('Int64')
    return data
```

Based on the survey's 100-point scoring criteria, any element presented in'interview will be considered essential and scored as 100. Elements not presented will be deemed not required and scored as 0. In the survey, missing values indicate the absence of any value, we need to calculate the average for each element. This is because, in the survey, the number of participants scoring each element is different.

```python
def sub_weight_num(intv_data,surv_data,col):

    elem_num = intv_data.shape[0] + surv_data.count()
    intv_count = count_key(intv_data,col)
    intv_score = pd.Series({key.strip():value for key, value in intv_count.items()})*100
    surv_score = surv_data.sum()
    total_score = surv_score.add(intv_score,fill_value=0).astype('Int64')
    elem_mean = total_score/elem_num
    weight_num = dict(round(elem_mean/elem_mean.sum(),3))
    
    return weight_num
    
```
